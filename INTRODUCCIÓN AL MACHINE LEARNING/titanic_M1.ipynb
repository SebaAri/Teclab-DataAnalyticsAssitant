```python
# leer el archivo de Excel
titanic = pd.read_excel("titanic_M1.xlsx")
titanic.head
```

```python
# a) Describir brevemente la cantidad de valores faltantes para cada una de las variables de la base de datos
titanic.isna()
```

```python
# Determinamos si hay datos faltantes
titanic.isna().sum()
```

```python
# b) En la columna Pclass, completar los valores NA, asignando a las personas a la clase 2
titanic1 = titanic.dropna()
titanic["pclass"] = titanic["pclass"].fillna(2)
# nos fijamos si hay valores faltantes en la columna pclass
titanic.isna().sum()
```

```python
# c) En la clase “Sex”, imputar los datos de manera de poder completar la base solo con los valores “male” y “female”

# Mostramos los distintos valores que tiene la columna sex
titanic.groupby("sex").size()
```

```python
# pasamos Hombre a male y Mujer a female
titanic["sex"] = np.where(titanic["sex"] == "Hombre", "male", titanic["sex"])
titanic["sex"] = np.where(titanic["sex"] == "Mujer", "female", titanic["sex"])
```

```python
# Vemos como quedó la columna sex
titanic.groupby("sex").size()
```

```python
# d) En los casos de las personas de las cuales no se tenga la edad, rellenar los datos con el promedio de edad de aquellos que sí tienen el dato

# Calculamos el promedio
mean = titanic["age"].mean()
mean
```

```python
# Rellenamos los datos de las personas que no tengan la edad con el promedio
titanic["age"] = titanic["age"].fillna(mean)
titanic
```

```python
# nos fijamos si en la edad hay algún dato vacío
titanic.isna().sum()
```

```python
# En la operación anterior nos figura que en la columna embarked hay 2 valores vacíos

# e) En la columna embarked, asignar el valor “S” a aquellos datos que se encuentran con valor vacío
titanic["embarked"] = titanic["embarked"].fillna("S")
```

```python
# nos fijamos si en la columna embarked quedó algún dato vacío
titanic.isna().sum()
```
